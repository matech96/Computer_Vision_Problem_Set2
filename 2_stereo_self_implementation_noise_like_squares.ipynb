```python
def disparity_ssd(L, R, kernel_size=5):
    size_x, size_y = L.shape
    depht_image = np.zeros(L.shape)
    for y in range(size_y-kernel_size):
        for x in range(size_x-kernel_size):
            kernel_L = L[y:y+kernel_size, x:x+kernel_size]
            min_diff = 256
            min_i = 0
            for i in range(size_x-kernel_size):
                kernel_R = R[y:y+kernel_size, x:x+kernel_size]
                squers = (kernel_L - kernel_R) ** 2
                diff = np.sum(squers)
                if diff < min_diff:
                    min_diff = diff
                    min_i = i
                depht_image[y, x] = min_i
    return depht_image
```

```python
kernel_size = 5
disparity = disparity_ssd(left, right, kernel_size)
plt.imshow(disparity,'gray')
plt.show()
```
