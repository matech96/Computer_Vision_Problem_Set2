```python
import cv2
print(cv2.__version__)
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
left = cv2.imread("input/pair1-L.png", flags=0)
right = cv2.imread("input/pair1-R.png", flags=0)
size_x, size_y = left.shape

print(type(left))
print(left.dtype)
figsize = 16
fig = plt.figure(figsize=(figsize,figsize))
r = fig.add_subplot(1,2,1)
r.set_title(left.shape)
plt.imshow(left, cmap="gray")
r = fig.add_subplot(1,2,2)
r.set_title(right.shape)
plt.imshow(right, cmap="gray")
plt.show()
```

```python
kernel_size = 17
numDisparities = 5
stereo = cv2.StereoBM_create(numDisparities=numDisparities*16, blockSize=kernel_size)
disparity = stereo.compute(left, right)


ground_truth = cv2.imread("input/pair1-D_L.png", flags=0)

fig = plt.figure(figsize=(figsize,figsize))
r = fig.add_subplot(1,2,1)
r.set_title("calculated")
plt.imshow(disparity, cmap="gray")
r = fig.add_subplot(1,2,2)
r.set_title("groud truth")
plt.imshow(ground_truth, cmap="gray")
plt.show()
```
